```python
import pandas as pd
```

```python
results = []
with open('data/labelled-tokens.txt', 'r') as tokens:
    for line in tokens:
        split_token = line.split('\t')
        results.append({
            "token": split_token[0].strip(),
            "code": split_token[1],
            "canonical": split_token[2].strip()
        })
        
df = pd.DataFrame(results)
```

```python
iv_token = df.query("code=='IV'").token.count()
oov_token = df.query("code=='OOV'").token.count()
no_token = df.query("code=='NO'").token.count()

iv_token, oov_token, no_token
```
